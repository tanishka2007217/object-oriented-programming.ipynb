```python
# super
```

```python
# Super() is used to access parent class.
```

```python
class parent:
    def __init__(self):
        print("Parent")
class child(parent):
    def __init__(self):
        super().__init__()
        print("child")
```

```python
child()
```

```python
# Method overriding
```

```python
# Method overriding replaces a parent method in the child class.
```

```python
class vehicle:
   def start(self):
       print("vehicle is starting...")
class car(vehicle):
    def start(self):
       print("car is starting...")
```

```python
car = car()
```

```python
car.start()
```

```python
# Using super
```

```python
class vehicle:
   def start(self):
       print("vehicle is stsrting...")
class car(vehicle):
   def start(self):
       super().start()   # call parent method
       print("car is starting...")
```

```python
car = car()
```

```python
car.start()
```

```python
# Abstract classes
```

```python
# ABC is a class used to force child classes to implement certain methods.
# @abstractionmethod marks methods that child classes must provide.
```

```python
# An abstract class is a class that cannot be instantiated and is used as ablueprint for other classes.
```

```python
from abc import ABC, abstractmethod

class shape(ABC):

  @abstractmethod
  def area(self):
    pass
class circle(shape):
  def area(self):
      return 78.5
```

```python
circle().area()
```

```python
from abc import ABC, abstractmethod
class calculator(ABC):
  @abstractmethod
  def calculate(self):
      pass
class Addition(calculator):
  def __init__(self,a,b):
    self,a = a
    self,b = b
  def calculate(self):
    return self.a + self.b
```

```python
add = Addition(10, 20)
```
